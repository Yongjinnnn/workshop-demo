```python
import matplotlib.pyplot as plt
```

```python
plt.hist(pokemon.hp,bins = 25)
plt.xlabel('HP')
plt.ylabel('Number of Pokemon')
plt.title ('distribution of HP over all pokemon')
```

```python
pokemon_groupies = pokemon.groupby('type1')
```

```python
water_pokemon = pokemon_groupies.get_group('water')
fire_pokemon = pokemon_groupies.get_group('fire')
```

```python
type(water_pokemon)
```

```python
plt.hist(water_pokemon.hp, bins=25, label = 'water')
plt.hist(fire_pokemon.hp, bins=25, label = 'fire')
plt.legend()
plt.title("Distribution of HP by type")
plt.xlabel("HP")
plt.ylabel("Number of Pokemon")
```
